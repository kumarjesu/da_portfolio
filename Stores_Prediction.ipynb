```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(color_codes=True)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('airlines_final.csv')
df.head()
```

```python
# To include NaN in the count, use df.nunique(dropna=False).
df.nunique()
```

```python
# Drop the last 1 columns
df = df.iloc[:, 1:]
df.head()
```

```python
#Check the number of unique value from all of the object datatype
df.select_dtypes(include='object').nunique()
```

```python
# list of categorical variables to plot
cat_vars = ['day', 'cleanliness', 'safety',
            'satisfaction', 'day','airline']

# create figure with subplots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15, 15))
axs = axs.flatten()


# create barplot for each categorical variable
for i, var in enumerate(cat_vars):
    sns.countplot(x=var, hue='day', data=df, ax=axs[i])
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=90)


# adjust spacing between subplots
fig.tight_layout()

# remove the sixth subplot
#fig.delaxes(axs[5])

# show plot
plt.show()
```
